## 문제1
- 주어진 데이터셋(members.csv)의 'views' 컬럼 상위 10개 데이터를 상위 10번째 값으로 대체한 후 'age'컬럼에서 80 이상인 데이터의 'views' 컬럼 평균값 구하기


```python
# your code
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/members.csv")
df
```

```python
df = df.sort_values('views', ascending=False)
```

```python
tenth = df['views'].iloc[9]
for i in range(0, 10):
    df['views'].iloc[i] = tenth
```

```python
cond = df['age'] >= 80
df[cond]['views'].mean()
```

## 문제3
- 주어진 데이터셋(members.csv)의 'age'컬럼의 이상치를 모두 더하시오!
- 단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함

```python
# 표준편차, 평균값 구하기
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/members.csv")
df
```

```python
strange = df['age'].std() * 1.5
avg = df['age'].mean()
first = avg - strange
second = avg + strange
```

```python
cond = (df['age']>second) | (df['age']<first)
df[cond]['age'].sum()
```

```python
# 473.5
```
