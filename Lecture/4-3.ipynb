```python
# 5
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p1/members.csv")
r1 = df['views'].quantile(.25)
r2 = df['views'].quantile(.75)
iqr = r2-r1
iqr
line1 = r1 - iqr* 1.5
line2 = r2 + iqr* 1.5
print(sum(df['views']>line2) + sum(df['views']<line1))
```
