## 스케일링 방법1

```python
# MinMaxScaler 활용
data = pd.read_csv("mtcars.csv")

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
print("MinMaxScaler 변환 전: \n", data['qsec'].head())
data['qsec'] = scaler.fit_transform(data[['qsec']])
print("MinMaxScaler 변환 전: \n", data['qsec'].head())
```

## 스케일링 방법2

```python
# minmax_scale 활용
data = pd.read_csv("mtcars.csv")

from sklearn.preprocessing import minmax_scale
print("MinMaxScaler 변환 전: \n", data['qsec'].head())
data['qsec'] = minmax_scale(data['qsec'])
print("MinMaxScaler 변환 전: \n", data['qsec'].head())
```

## 스케일링 방법3

```python
# min-max 함수
data = pd.read_csv("mtcars.csv")

def my_minmax(data):
    data = (data-min(data)) / (max(data) - min(data))
    return data

print("MinMaxScaler 변환 전: \n", data['qsec'].head())
data['qsec'] = my_minmax(data['qsec'])
print("MinMaxScaler 변환 전: \n", data['qsec'].head())
```

## 조건(0.5보다 큰 값)

### sum 활용

```python
cond = data['qsec'] > 0.5
print(sum(cond))
```

### len 활용

```python
cond = data['qsec'] > 0.5
print(len(data[cond]))
```

### count 활용

```python
cond = data['qsec'] > 0.5
print(data[cond]['qsec'].count())
```
