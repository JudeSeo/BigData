```python
# 1
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p1/members.csv")

df['f1'] = df['f1'].fillna(df['f1'].median())
df=df.dropna(axis=0)
df=df[:int(len(df)*0.7)]
print(df['views'].quantile(.75)-df['views'].quantile(.25))
```

```python
# 2
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p1/members.csv")
df = df.dropna(subset=['f1']) # 특정 컬럼의 결측치 제거
df['f3']=df['f3'].fillna(df['f3'].mode()[0]) # 특정 컬럼의 결측치 최빈값 대체
print(sum(df['f3'] == 'gold'))
# print(len(df[df['f3'] == 'gold']))
```

```python
# 3 replace
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p1/members.csv")
df = df.dropna(subset=['views']) # 특정 컬럼의 결측치 제거
df['f3'] = df['f3'].fillna('0')
df['f3'] = df['f3'].str.replace('silver', '1')
df['f3'] = df['f3'].str.replace('gold', '2')
df['f3'] = df['f3'].str.replace('vip', '3')
df['f3']=df['f3'].astype(int)

print(df['f3'].sum())
```

```python
# 3 map
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p1/members.csv")
df = df.dropna(subset=['views']) # 특정 컬럼의 결측치 제거

import numpy as np
dict_list = {np.nan:0, 'silver':1, 'gold':2, 'vip':3}
df['f3'] = df['f3'].map(dict_list)

print(df['f3'].sum())
```
