```python
import pandas as pd

train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p2/ab_nyc/train.csv")
test =  pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p2/ab_nyc/test.csv")
print(train.shape, test.shape)
```

```python
cols = ['name', 'host_name', 'last_review', 'host_id']
print(train.shape)
train = train.drop(cols, axis=1)
test = test.drop(cols, axis=1)
print(train.shape)
```

```python
train['reviews_per_month'] = train['reviews_per_month'].fillna(0)
test['reviews_per_month'] = test['reviews_per_month'].fillna(0)
train.isnull().sum()
test.isnull().sum()
```

```python
train = train.drop('id', axis=1)
test_id = test.pop('id')
test.head()
```

```python
train.info()
```

```python
cols =['neighbourhood_group', 'neighbourhood', 'room_type']
from sklearn.preprocessing import LabelEncoder

for col in cols:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])

train[cols]
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(train.drop('price', axis=1), train['price'], test_size=0.15, random_state=2022)
X_tr.head()
```

```python
import numpy as np
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(y_test, y_pred))

def rmsle(y_test, y_pred):
    return np.sqrt(np.mean(np.power(np.log1p(y_test) - np.log1p( y_pred), 2)))

def mape(y_test, y_pred):
    return np.mean(np.abs((y_test - y_pred) / y_test)) * 100
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression();
model.fit(X_tr, y_tr);
pred = model.predict(X_val)
print("r2: ", r2_score(y_val, pred))
print("mae: ", mean_absolute_error(y_val, pred))
print("mse: ", mean_squared_error(y_val, pred))
print("rmse: ", rmse(y_val, pred))
print("rmsle: ", rmsle(y_val, pred))
print("mape: ", mape(y_val, pred))

# r2:  0.08452239774949499
# mae:  73.98120907242603
# mse:  44813.521129753695
# rmse:  211.6920431422818
# rmsle:  0.6284667216269796
# mape:  58.89870347352111
```

```python
from sklearn.linear_model import Ridge

model = Ridge();
model.fit(X_tr, y_tr);
pred = model.predict(X_val)
print("r2: ", r2_score(y_val, pred))
print("mae: ", mean_absolute_error(y_val, pred))
print("mse: ", mean_squared_error(y_val, pred))
print("rmse: ", rmse(y_val, pred))
print("rmsle: ", rmsle(y_val, pred))
print("mape: ", mape(y_val, pred))

# r2:  0.08458002418498123
# mae:  73.94214514943883
# mse:  44810.700259556594
# rmse:  211.68538036330378
# rmsle:  0.6215134552796238
# mape:  58.858160278049674
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor();
model.fit(X_tr, y_tr);
pred = model.predict(X_val)
print("r2: ", r2_score(y_val, pred))
print("mae: ", mean_absolute_error(y_val, pred))
print("mse: ", mean_squared_error(y_val, pred))
print("rmse: ", rmse(y_val, pred))
print("rmsle: ", rmsle(y_val, pred))
print("mape: ", mape(y_val, pred))

# r2:  0.220036157766756
# mae:  65.79755140877073
# mse:  38179.99046447351
# rmse:  195.3970073068508
# rmsle:  0.4915272817139663
# mape:  45.39337302853444
```
