## 문제1
- 결측치 데이터(행)을 제거하고, 앞에서부터 70% 데이터만 활용해, 'f1' 컬럼 1사분위 값을 구하시오
- 데이터: members.csv

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/members.csv")
df
```

```python
# 결측치 제거
df.isnull().sum()
df = df.dropna()
```

```python
# 앞에서부터 70% 데이터
cnt = int(len(df) * 0.7)
df = df.iloc[:cnt]
```

```python
# f1컬럼의 1사분위 값
df['f1'].quantile(0.25)
```

```python
# 57.0
```

## 문제2
- index는 년도임. 2000년 데이터 중 2000년 평균보다 큰 값의 데이터 수
- 데이터: 아래 셀 실행

```python
# 데이터 생성(먼저 실행해 주세요)
import pandas as pd
import random
random.seed(2022)
df = pd.DataFrame()
for i in range(0, 3):
    list_box = []
    for k in range(0, 200):
        ran_num = random.randint(1,200)
        list_box.append(ran_num)
    df[i+1999] = list_box
df = df.T
df
```

```python
mean = df.loc[2000].mean()
cond = df.loc[2000] > mean
len(df.loc[2000][cond])
```

```python
# 100
```
