```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/8_2/churn_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/8_2/churn_test.csv")
```

```python
print(train.shape)
print(test.shape)
print(train.info())
y_train = train.pop('TotalCharges')
print(train.info())
```

```python
from sklearn.preprocessing import LabelEncoder
cols = train.select_dtypes('object').columns

print(train.info())
print(test.info())
for col in cols:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col] = le.fit_transform(test[col])

print(train.info())
print(test.info())
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)
print(X_tr.shape, X_val.shape, y_tr.shape, y_val.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_tr, y_tr)
pred = model.predict(X_val)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_val, pred)
```

```python
pred = model.predict(test)

submit = pd.DataFrame({
    'pred':pred
}).to_csv("result.csv", index=False)

df = pd.read_csv("result.csv")
print(df)
```
