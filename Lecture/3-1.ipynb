```python
# 데이터 프레임 만들기 (할인율과 칼로리 -> 문자열)
data = {
    "메뉴":['아메리카노', '카페라떼', '카페모카', '바닐라콜드브루'],
    "가격":[4100, 4600, 4600, 5100],
    "할인율":['0.5', '0.1', '0.2', '0.3'],
    "칼로리":[10,180,420,320],
}
df = pd.DataFrame(data)
df.info()
```

```python
# 자료형 변환 / astype / object -> float
df['할인율'] = df['할인율'].astype('float')
```

```python
# 자료형 확인
df.info()
```

```python
# 새로운 컬럼 추가 (값은 0)
df['new'] = 0;
df
```

```python
# 할인가 컬럼 추가
df['할인가'] = df['가격'] * (1-df['할인율'])
df
```

```python
# 결측값으로 추가, 원두 컬럼을 만들거 결측값(NaN)으로 대입
import numpy as np
df['원두'] = np.nan
df
```

```python
# 데이터 삭제
# axis=1: 열방향(컬럼) / axis=: 행방향
# 코랩 판다스 버전이 올라가면서 워닝이 발생함. 따라서 axis사용 권장함
df = df.drop('new', axis=1)
```

```python
# 데이터 확인
df.head(2)
```

```python
# 행 삭제
df = df.drop(1, axis=0)
```

```python
# 데이터 확인
df.head(2)
```
