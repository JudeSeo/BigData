```python
# 리스트에 있는 값, 반복문을 통해 출력
listbox = [2,4,6,8,10]
for item in listbox:
    print(item)
```

```python
# 리스트에 있는 값 반복문을 통해 +1 한 값, 출력
listbox = [2,4,6,8,10]
for item in listbox:
    # print(item+1)
    result = item+1
    print(result)
```

```python
# 리스트에 있는 문자열 출력
listbox = ['네모','세모','동그라미']
for item in listbox:
    print(item)
```

```python
# 리스트에 있는 문자열과 인덱스값 출력
listbox = ['네모','세모','동그라미', '별']
for index, item in enumerate(listbox):
    print(index)
    print(item)
```

```python
# Tip (print(f스트))
for index, item in enumerate(listbox):
    print(f'인덱스{index}인 값은{item}입니다')
```

```python
# 함수 정의 (리턴값)
def plus(x, y):
    result = x+y
    return result
```

```python
# 함수 호출
a=plus(2, 3)
print(a)
```

```python
# 리턴값이 2개 (최소, 최대값을 구하는 함수)
listbox = [15, 46, 78, 24, 56]
def min_max(data):
    mi = min(data)
    ma = max(data)
    return mi, ma

a, b = min_max(listbox)
print(a, b)
```

```python
# 평균을 구하는 함수
listbox = [15, 46, 78, 24, 56]

def mean(data):
    return sum(data) / len(data)

mean(listbox)
```

```python
# 주어진 리스트의 최대값-최소값을 구하는 함수를 만드시오.
listbox = [15, 46, 78, 24, 56]

def maxmin(data):
    return max(data) - min(data)

maxmin(listbox)
```
