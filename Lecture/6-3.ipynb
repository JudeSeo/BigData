```python
import pandas as pd

train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p2/heart/2files/train.csv")
test =  pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p2/heart/2files/test.csv")
```

```python
train = train.drop('id', axis=1)
test_id = test.pop('id')

test.head()
```

```python
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train.drop('output', axis=1), train['output'], test_size=0.15, random_state=2022)
```

```python
X_tr.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score

model = RandomForestClassifier(random_state=2022, max_depth=5, n_estimators=400)
model.fit(X_tr, y_tr)
pred = model.predict(X_val)
pred_proba = model.predict_proba(X_val)

print(roc_auc_score(y_val, pred_proba[:,1]))
print(f1_score(y_val, pred))
print(accuracy_score(y_val, pred))

# 0.9301242236024845
# 0.7906976744186046
# 0.7567567567567568

# max_depth=3
# 0.9378881987577639
# 0.8636363636363636
# 0.8378378378378378

# max_depth=5
# 0.9409937888198757
# 0.8444444444444444
# 0.8108108108108109

# max_depth=7
# 0.9347826086956521
# 0.8444444444444444
# 0.8108108108108109

# max_depth=5, n_estimators=200
# 0.9440993788819876
# 0.8444444444444444
# 0.8108108108108109

# max_depth=5, n_estimators=400
# 0.9409937888198757
# 0.8444444444444444
# 0.8108108108108109
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(random_state=2022, max_depth=5, n_estimators=400, learning_rate=0.01)
model.fit(X_tr, y_tr)
pred = model.predict(X_val)
pred_proba = model.predict_proba(X_val)

print(roc_auc_score(y_val, pred_proba[:,1]))
print(f1_score(y_val, pred))
print(accuracy_score(y_val, pred))

# 0.9192546583850931
# 0.8444444444444444
# 0.8108108108108109

# max_depth=3
# 0.8975155279503105
# 0.8444444444444444
# 0.8108108108108109

# max_depth=5
# 0.906832298136646
# 0.8181818181818182
# 0.7837837837837838

# max_depth=5, n_estimators=200
# 0.9068322981366459
# 0.8181818181818182
# 0.7837837837837838

# n_estimators=400, learning_rate=0.01
# 0.9161490683229814
# 0.8372093023255814
# 0.8108108108108109
```
