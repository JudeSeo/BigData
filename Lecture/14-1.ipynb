## 작업형1 문제
1-1. age 컬럼의 3사분위수와 1사분위수의 차를 절대값으로 구하고, 소수점 버려서, 정수로 출력

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/4_1/basic1.csv")
df
```

```python
# age 컬럼의 3사분위수
up = df['age'].quantile(0.75)
```

```python
# age 컬럼의 1사분위수
down = df['age'].quantile(0.25)
```

```python
# 차를 절대값으로 구해 소수점 버려서 출력
int(abs(up-down))
```

```python
# 50
```

1-2.(loves반응+wows반응)/(reactions반응) 비율이 0.4보다 크고 0.5보다 작으면서, type 컬럼이 'video'인 데이터의 갯수

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/4_1/fb.csv")
df
```

```python
# (loves반응+wows반응)/(reactions반응) 비율
df['rate'] = (df['loves']+df['wows'])/df['reactions']
df
```

```python
# 비율이 0.4보다 크고 0.5보다 작은 데이터의 갯수
cond1 = (df['rate'] > 0.4) & (df['rate'] < 0.5)
```

```python
# type 컬럼이 'video'인 데이터의 갯수
cond2 = df['type'] == 'video'
```

```python
len(df[cond1 & cond2])
```

```python
# 90
```

1-3. date_added가 2018년 1월 이면서 country가 United Kingdom 단독 제작인 데이터의 갯수

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/4_1/nf.csv")
df
```

```python
# date_added가 2018년 1월
df['date_added'] = pd.to_datetime(df['date_added'])
df['year'] = df['date_added'].dt.year
df['month'] = df['date_added'].dt.month
cond1 = (df['year'] == 2018) & (df['month'] == 1)
```

```python
# country가 United Kingdom 단독
cond2 = df['country'] == 'United Kingdom'
```

```python
len(df[cond1 & cond2])
```
