```python
import pandas as pd

data = {
    '상품명': ['크리스마스트리', '산타모자', '선물상자', '장식품', '캔디케인', '산타양말', '눈사람모형'],
    '가격': [50, 10, 20, 15, 5, 12, 30],
    '판매량': [200, 500, 300, 400, 1000, 300, 150],
    '카테고리': ['장식', '의류', '포장', '장식', '간식', '의류', '장식']
}

df = pd.DataFrame(data)
```
