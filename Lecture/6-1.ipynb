```python
import pandas as pd

train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p2/ceredit%20card/train.csv")
test =  pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p2/ceredit%20card/test.csv")
print(train.shape, test.shape)
```

```python
# baseline
cols = train.select_dtypes(include='object').columns
cols

# print(train.shape, test.shape)
# # train = train.drop(cols, axis=1)
# train.drop(cols, axis=1, inplace=True)
# test.drop(cols, axis=1, inplace=True)
# print(train.shape, test.shape)

# label
from sklearn.preprocessing import LabelEncoder
for col in cols:
    le=LabelEncoder()
    train[col]=le.fit_transform(train[col])
    test[col]=le.transform(test[col])
train[cols].head()

# one-hot
# train = pd.get_dummies(train, columns=cols)
# test = pd.get_dummies(test, columns=cols)
```

```python
train.drop("CLIENTNUM", axis=1, inplace=True)
test_id = test.pop("CLIENTNUM")
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(train.drop("Attrition_Flag", axis=1), train['Attrition_Flag'], test_size=0.2, random_state=2022 )
```

```python
X_tr.head()
```

```python
# ëª¨ë¸ë¶ˆëŸ¬ì˜¤ê¸°
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=2022)
# í•™ìŠµ
model.fit(X_tr, y_tr)
#ì˜ˆì¸¡
pred=model.predict(X_val)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# ì •í™•ë„
print(accuracy_score(y_val, pred))

# ì •ë°€ë„
print(precision_score(y_val, pred))

# ìž¬í˜„ìœ¨
print(recall_score(y_val, pred))

# f1
print(f1_score(y_val, pred))

# baseline
# 0.9666872301048736
# 0.9444444444444444
# 0.8435114503816794
# 0.8911290322580645

# label í›„
# 0.9685379395434917
# 0.948936170212766
# 0.851145038167939
# 0.89738430583501

# one-hot í›„
# 0.9605181986428131
# 0.9419642857142857
# 0.8053435114503816
# 0.8683127572016461
```

```python
# roc_auc
pred_proba = model.predict_proba(X_val)
print(roc_auc_score(y_val, pred_proba[:,1]))

# baseline
# 0.9894048160692921

# label í›„
# 0.9899707351049547

# one-hot í›„
# 0.9885945548197205
```
