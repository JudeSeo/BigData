```python
# 학습 전 실행: csv파일 (data.csv) 생성
import pandas as pd
import numpy as np
data = {
    "메뉴":['아메리카노','카페라떼','카페모카', '바닐라라떼', '녹차', '초코라떼', '바닐라콜드브루'],
    "가격":[4100, 4600, 4600, 5100, 4100, 5000, 5100],
    "할인율":[0.5, 0.1, 0.2, 0.3, 0, 0, 0],
    "칼로리":[10, 180, 420, 320, 20, 500, 400],
    "원두":['콜롬비아', np.NaN, '과테말라', np.NaN, '한국', '콜롬비아', np.NaN],
    "이벤트가":[1900, 2300, np.NaN, 2600, np.NaN, 3000, 3200],
}
data = pd.DataFrame(data)
data.to_csv('data.csv', index=False)
```

```python
# 데이터 불러오기
df = pd.read_csv('data.csv')
df
```

```python
# 1개 조건
# 할인율 > 0.2
# cond = df['할인율'] > 0.2
# df[cond]
df[df['할인율'] > 0.2]
```

```python
# 2개 이상일때 (AND)
# 할인율 >= 0.2
# 칼로리 < 400
cond1 = df['할인율'] >= 0.2
cond2 = df['칼로리'] < 400
df[cond1 & cond2]
```

```python
# 2개 이상일때 (OR)
# 할인율 >= 0.2
# 칼로리 < 400
cond1 = df['할인율'] >= 0.2
cond2 = df['칼로리'] < 400
df[cond1 | cond2]
```

```python
# 문자열 1개 조건
# 원두 == 콜롬비아
cond = df['원두'] == '콜롬비아'
df[cond]
```

```python
# 문자열과 숫자 조건 (AND)
# 원두 == 콜롬비아
# 가격 < 4500
cond1 = df['원두'] == '콜롬비아'
cond2 = df['가격'] < 4500
df[cond1 & cond2]
```

```python
# 결측치

# 컬럼별 결측치 확인
df.isnull().sum()
```

```python
# 결측값 채우기
# 원두 -> 코스타리카로 채우기
df['원두'] = df['원두'].fillna('코스타리카')
```

```python
# Quiz
# 데이터 불러오기
df = pd.read_csv('data.csv')
df
```

```python
# '이벤트가'컬럼 결측치는 이벤트 가격 데이터 중 최소값으로 결측치 채움
min_price = df['이벤트가'].min()
df['이벤트가'] = df['이벤트가'].fillna(min_price)
df
```

```python
#'원두'컬럼 결측치는 원두 데이터 중 최빈값으로 결측치 채움
f=df['원두'].mode()[0]
df['원두'] = df['원두'].fillna(f)
df
```

```python
# 가격이 5000 이상인 데이터의 수를 구하시오.
cond = df['가격']>=5000
len(df[cond])
```
